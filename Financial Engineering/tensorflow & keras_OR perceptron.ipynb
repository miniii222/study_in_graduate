```python
import numpy as np
import tensorflow as tf
```

##### y = Xw+b

```python
w = tf.Variable(tf.zeros([2,1]))
b = tf.Variable(tf.zeros([1]))
```

```python
x = tf.placeholder(tf.float32, shape = [None,2])
#t : true
t = tf.placeholder(tf.float32, shape = [None, 1])
#y : y_hat
y = tf.sigmoid(tf.matmul(x,w)+b)
```

```python
#실제 데이터
X = np.array([[0,0], [1,0], [0,1],[1,1]])
Y = np.array([[0],[1],[1],[1]])
```

### loss function

```python
cross_entropy = -tf.reduce_sum(t * tf.log(y) + (1-t)*tf.log(1-y))
```

### optimizer

```python
train_step = tf.train.GradientDescentOptimizer(0.1).minimize(cross_entropy)
```

```python
correct_prediction = tf.equal( tf.to_float(tf.greater(y,0.5)), t) 
```

### variable initialization

```python
sess = tf.Session()
init = tf.global_variables_initializer()
sess.run(init)
```

```python
for epoch in range(2000) :
    sess.run(train_step, feed_dict = {
        x: X,
        t: Y
    })
    
    my_c_e = sess.run(cross_entropy, feed_dict = {
        x:X, t:Y
    })
    if epoch % 10 ==0 :
        print(-my_c_e)
```

```python
sess.run(w)
```

```python
sess.run(y, feed_dict = {x: X, t:Y})
```

```python
correct_prediction.eval(session = sess, feed_dict = {
    x:X, 
    t:Y
})
```

```python
sess.run(correct_prediction, feed_dict = {
    x:X,
    t:Y
})
```

```python
accuracy = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))
accuracy = accuracy.eval(session =sess, feed_dict = {x:X, t:Y})
accuracy
```

```python
import numpy as np
from keras.models import Sequential
from keras.layers import Dense, Activation
from keras.optimizers import SGD
```

```python
model = Sequential([
    
    Dense(input_dim = 2, units = 1), #XW+b
    Activation('sigmoid')
]
    
)
```

```python
model.compile(loss = 'binary_crossentropy', optimizer=SGD(lr = 0.1))
```

```python
X = np.array([[0,0], [1,0], [0,1],[1,1]])
Y = np.array([[0],[1],[1],[1]])
```

```python
model.fit(X,Y, epochs=200, batch_size=4, verbose=1)
```

```python
model.fit(X,Y, epochs=1000, batch_size=4, verbose=0)
```
